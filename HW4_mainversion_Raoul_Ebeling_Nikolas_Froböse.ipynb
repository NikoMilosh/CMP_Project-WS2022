# Homework 4. Moleculal dynamics simulation  of 2D soft-disk fluid

In this homework, you will use molecular dynamics to simulate the behavior of fluid consisting of soft disks in two dimensions. The particles in this fluid interact with each other via the pairwise potential of the following form: 

 $$  u(r_{ij})=\left\{
                \begin{array}{ll}
                  4\epsilon \left[ 
                  \left (
                  \frac{\sigma}{r_{ij} }
                  \right )^{12} 
                  -
                   \left (
                  \frac{\sigma}{r_{ij} }
                  \right )^{6} + 
                  \right ] + \epsilon,  \ r_{ij} < r_c\\
                  0, r_{ij} \geq r_c, r_c = 2^{1/6}\sigma
                \end{array}
              \right.
$$

For this model, we will use reduced MD units, such that $\sigma = 1 $, $  \epsilon=1$,  $k_b = 1$.
All the particles have unit mass.

$ r_c = 2^{1/6}\sigma $

1) Initialization of the grid

```python
def init_lattice(N_particles, L):        # Refernece: https://www.kite.com/python/answers/how-to-get-all-element-combinations-of-two-numpy-arrays-in-python

    '''
    
    N_particles... number of particels inside the fluid. It needs to be a square number, 
    else one can not build an equally distance square lattice.
    
    L... length of the box.
    
    returns: 2-D Array of grid particles, vectorized.
    
    '''
    
    nps = np.sqrt(N_particles)    #Calculates how many particles are in one line
    
    distance = L/(nps*2) #we parametrize the particles to be within the length L
    
    init=np.linspace(distance,L-distance,int(nps))           #Generates an array with the number of particles in a line with equal distances   
    
    grid_points=np.array(np.meshgrid(init,init)).T.reshape(-1, 2)  #
    grid_points=grid_points-L/2
    
    
    return grid_points  #returns grid 
```

2) Initialize velocity

```python
def init_velocity(T,N):
    
    '''
    
    N_particles... number of particels inside the fluid. It needs to be a square number, 
    else one can not build an equally distance square lattice.
    
    T - Temperature for the Normal distribution
    
    returns: 2-D Array of grid particles, vectorized.
    
    '''
    
    
    v= np.random.normal(0,np.sqrt(T),(N,2)) #Normal distribution
    
    v = v - np.mean(v,axis=0)
    return v
```

```python
def min_image(r_int,index): # Improve by vectorizing for loop
    
    '''
    
    
    
    '''
    
    r = r_int.copy()
    x = r[:,:1]
    y = r[:,1:]    
    for i in range(len(x)):
        if x[i] > L/2+r[index,0]:
            x[i] = x[i] - L
        elif x[i] < -L/2+r[index,0]:
            x[i] = x[i] + L
        else:
            x[i] = x[i]
        
    # updating y
    for i in range(len(y)):
        if y[i] > L/2+r[index,1]:
            y[i] = y[i] - L
        elif y[i] < -L/2+r[index,1]:
            y[i] = y[i] + L
        else:
            y[i] = y[i]
    pos_new = np.concatenate((x,y),axis=1)    
    return pos_new  
```

```python
def LJ_trunc(N,position,L):
    
    '''
    N - Number of particles
    
    position - 2D array (x,y-coordinates) of the particle positions
  
    
    L - length of the box.
    
    position - 2 D array of x and y coordinates
    
    
    Returns: Lenard-Jones- force between the particles
    
    '''
    
    
    forces = np.zeros((N, 2))    #Setting list and dimension for forces
    
    rc_squared = 2**(1/3)       #Cut-off radius
    
    #Loop over particles to calculate distances (according to minimal image) 
    #to then calculate the energies
    
    for i in range(N):
        
        force = np.zeros((N-i-1,2)) #initialized force
        
        pos = min_image(position,i)   #determine minimal image
        
        r_vec = pos[i]-pos[i + 1:]      #vector between two particle positions
        
        r_squared = np.sum(r_vec ** 2, axis=1)           # calculates [1/r12^2,1/r13^2,...]. Is like [:, np.newaxis] reshape(143,1)
        
        index = np.where(r_squared<rc_squared)   #finds particles that are within the cut off radius
        
        r_squared_new = r_squared[index][:, np.newaxis]
        r_vec_new = r_vec[index,:]                                       
        
        force[index,:] = 4*(12*1/r_squared_new**7 - 6*1/r_squared_new**4)*r_vec_new  # calculates force arising from LJ-potential
        forces[i] += np.sum(force, axis=0)                                # sums them up
        forces[i + 1:] -= force                                           # for next loop F21 is missing, but F21 =-F12 and we save is preventively
    return forces                                                       # F=a if m=1
```

```python
def apply_pbc(L, position): # Improve by vectorizing for loop
    
    '''
    position - 2D array (x,y-coordinates) of the particle positions
    
    position - 2 D array of x and y coordinates
    
    Returns: Lenard-Jones- force between the particles
    
    '''
    
    #Initializing x and y coordinates
    
    x = position[:,:1]
    y = position[:,1:]
    
    # updating x
    for i in range(len(x)):
        if x[i] > L/2:
            x[i] = x[i] - L
        elif x[i] < -L/2:
            x[i] = x[i] + L
        else:
            x[i] = x[i]
        
    # updating y
    for i in range(len(y)):
        if y[i] > L/2:
            y[i] = y[i] - L
        elif y[i] < -L/2:
            y[i] = y[i] + L
        else:
            y[i] = y[i]
        
    pos_new = np.concatenate((x,y),axis=1)   #Returns a 2D array instead of 2 1D-arrays
    return pos_new
```

```python
def valet_velocity(x_cur,v_cur,N,L,dt,force_cur,force):
    
    '''
    x_cur - current position
    
    v_cur - current velocity
    
    N - Particle number
    
    L - Box size
    
    dt - integration step (time step)
    
    force_cur - current corce
    
    force - type of force: in this project LJ_trunc()
    
    Returns:
    
    x_new - updated positions (2-D array)
    
    v_new - updated velocity (2-D array)
    
    force_new - updated force (2-D array)
    
    
    
    '''
    
    x_new = x_cur+v_cur*dt+1/2*force_cur*dt**2 #new positions
    
    x_new = apply_pbc(L,x_new) #periodic boundry conditions
    
    force_new = force(N,x_new,L) #force
    
    v_new = v_cur+dt/2*(force_cur+force_new)
    
    return x_new, v_new, force_new
```

```python
def md(N,L,x0,v0,force_cal,boundary,integrator,dt,nsteps):
    
    '''
    N - Particle number
    
    L - Box size
    
    x0 - initial positions
    
    v0 - initial velocity
    
    force_cal - calculated force
    
    boundary - boundary condition
    
    integrator - the integrator one will use, e.g. Verlet, Euler, Langevine,...
    
    dt - integration step
    
    nsteps - How many steps do we want to run the simulation?
    
    
    Returns: trajectory in x-y-plane (2D-array), velocity of x-y-plane at each step (2D array)
    
    '''
    
    x = np.zeros((nsteps+1,N,2))
    v = np.zeros((nsteps+1,N,2))
    
    x[0]=x0
    v[0]=v0
    
    force_cur = force_cal(N,x0,L)
    
    for i in tqdm(range(nsteps)):    
        x[i+1], v[i+1], force_new = integrator(x[i],v[i],N,L,dt,force_cur,force_cal)
        force_cur = force_new
    return x, v
```

```python
def energy_velocities(v,r,N):
    
    '''
    N - Particle number
    
    v - 2D array of velocity in x-y-plane
    
    r - distances between particles for potential calculations
    
    
    Returns: 
    
    valocity (each coordinate as 1D array)
    
    E_kin - kinetic energy
    
    E_pot - Potential energy 
    
    E_tot - total energy 
    
    tem - temperature
    
    '''
    
    v_step = np.sum(v,axis=1) #summation of velocities for E_kin
    
    #Initializing velocities
    
    vx = v_step[:,:1]
    vy = v_step[:,1:]
    
    E_kin = np.sum(1/2*np.sum(v**2, axis=1), axis=1)   #Kinetic energy
    
    rc = 2**(1/6)  #Cut-off radius
    
    E_pot_t = np.zeros(len(r))  #making a list for potential energies
    
    #loop over po
    
    for n in tqdm(range(len(r))):
        
        position=r[n]
        E_pot=0
        
        for i in range(len(position)):
            
            pos = min_image(position,i)
            rij = np.linalg.norm(pos[i]-pos[i + 1:], axis = 1)
            index = np.where(rij<rc)
            r_inv = rij[index]**-6
            E_pot+=np.sum(4*(r_inv**2-r_inv)+1)
            
        E_pot_t[n] = E_pot
    E_tot = E_kin+E_pot_t
    temp = E_kin/N
    
    return vx , vy , E_kin, E_pot_t, E_tot, temp
```

```python
def langevin(x_cur,v_m_half,N,L,dt,T ,force_cur,force,gamma=1.0):
    
    '''
    N - Particle number
    
    x_cur - current position
    
    force - initalized force at current position
    
    L - box size
    
    T - Temperature
    
    dt - time step
    
    
    
    Returns: 
    
        x_new - updated position (2D)
        
        v_new - updated velovity (2D)
        
        force_new - updated force between particles (2D)
        
        v_p_half - necesarry for leap frog algorithm 
        
    
    '''
    
    #Constants
    
    a = np.exp(-gamma*dt)
    R = np.random.normal(0,1,(N,2))
    
    #leap frog
    
    v_p_half = v_m_half*a+force_cur*(1-a)/gamma+np.sqrt(T*(1-a**2))*R
    
    x_new = x_cur+v_p_half*dt
    
    x_new = apply_pbc(L,x_new) #periodic boundary conditions
    
    force_new = force(N,x_new,L) #updated force
    
    v_new = (v_p_half+v_m_half)/2 #updated velocities
    
    
    return x_new,v_new ,v_p_half, force_new
```

```python
def md(N,L,x0,v0,T,force_cal,boundary,integrator,dt,nsteps):
    
    x = np.zeros((nsteps+1,N,2)) #position list to update
    
    v_half = np.zeros((nsteps+1,N,2))
    v_full = np.zeros((nsteps,N,2))
    
    #Initialize position, velocity and force
    
    x[0]=x0
    v_half[0]=v0
    force_cur = force_cal(N,x0,L)
    
    #run multiple steps to obtain movement
    
    for i in tqdm(range(nsteps)):    
        x[i+1] ,v_full[i],v_half[i+1], force_new = integrator(x[i],v_half[i],N,L,dt,T,force_cur,force_cal)
        force_cur = force_new
    return x, v_half ,v_full
```
